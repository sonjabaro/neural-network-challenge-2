## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
#evaluate the data for the attrition column based on question involving accuracy (futher down after running the models)
attrition_df["Attrition"].value_counts()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# I used the approach we learned in our activities which involved doing our preprocessing prior to splitting the data.
```

```python
#Preprocess Y
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
#encode y - Attrition and Department columns
enc_attrition = OneHotEncoder(sparse_output=False)
attrition_encoded = enc_attrition.fit_transform(attrition_df[["Attrition"]])
attrition_columns = enc_attrition.get_feature_names_out(["Attrition"])
df_attrition = pd.DataFrame(attrition_encoded, columns= attrition_columns)
```

```python
enc_department = OneHotEncoder(sparse_output=False)
department_encoded = enc_department.fit_transform(attrition_df[["Department"]])
department_columns = enc_department.get_feature_names_out(["Department"])
df_department = pd.DataFrame(department_encoded, columns= department_columns)
```

```python
#concat to original data frame - based on what we'd done previously I added the new columns to our dataframe and removed the originals
#Further work identified I didn't need the concated dataframe so it's not used other than creation of my y_df that is already encoded.
encoded_df = pd.concat([attrition_df,df_department,df_attrition], axis =1)
```

```python
#drop original "Attrition and Department" columns
encoded_df = encoded_df.drop(["Attrition","Department"], axis=1)
```

```python
# Create y_df with the Attrition and Department columns

y_attrition= encoded_df[["Attrition_No","Attrition_Yes"]]
y_department = encoded_df[["Department_Sales","Department_Human Resources", "Department_Research & Development"]]
#convert the data to an array
y_attrition = y_attrition.to_numpy()
y_department = y_department.to_numpy()
```

```python
# Create a list of at least 10 column names to use as X data
columns = attrition_df[['Age', 'DistanceFromHome', 'Education', 'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]


# Create X_df using your selected columns
X_df = pd.DataFrame(columns)

# Show the data types for X_df
print(X_df.dtypes)
X_df.head()
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_atttrition_train,y_attrition_test, y_department_train, y_department_test = train_test_split(X_df,y_attrition,y_department,random_state=78)
```

## Create, Compile, and Train the Model

```python
from tensorflow.keras.optimizers import Adam
```

```python
# Find the number of columns in the X training data
#Because I created a list of more than 10 features my output for numbers will be different from the example
num_cols = X_train.shape[1]
print(num_cols)

# Create the input layer
input_layer = layers.Input(shape = (X_train.shape[1],), name="input_features")

# Create at least two shared layers
shared_layer1=layers.Dense(64, activation="relu")(input_layer)
shared_layer2=layers.Dense(128, activation = "relu")(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
# Create the hidden layer
department_hidden = layers.Dense(units=32, activation='relu', name="department_hidden")(shared_layer2)

# Create the output layer
department_output = layers.Dense(units=3, activation='softmax', name="department_output")(department_hidden)

```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden = layers.Dense(units=32, activation='relu', name="attrition_hidden")(shared_layer2)

# Create the output layer
attrition_output = layers.Dense(units=2, activation='softmax', name="attrition_output")(attrition_hidden)
```

```python
from tensorflow.keras.optimizers import Adam
model = Model(inputs=input_layer, outputs=[attrition_output, department_output])
# Compile the model
model.compile(optimizer=Adam(),
              loss={"attrition_output": "binary_crossentropy", "department_output": "categorical_crossentropy"},
              metrics={"attrition_output": "accuracy", "department_output": "accuracy"})
# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_df,
          {"attrition_output":y_attrition, "department_output":y_department},
          epochs=100,
          batch_size = 32,
          validation_split = 0.2)
```

```python
# Evaluate the model with the testing data
test_results = model.evaluate(X_test, {"attrition_output": y_attrition_test, "department_output": y_department_test})
test_results
```

```python
# Print the accuracy for both department and attrition
print(f"Attrition Accuracy: {test_results[3]}")
print(f"Department Accuracy: {test_results[4]}")
```
